# The Intern

Talented young professionals starting their careers are now less restricted by national borders and may want to work in a location with high wages. But what if a location has an equally high cost of living? This notebook scrapes [net salary after tax](https://www.numbeo.com/cost-of-living/rankings.jsp) and [cost of living index](https://www.numbeo.com/cost-of-living/prices_by_city.jsp?displayCurrency=USD&itemId=105) data from [Numbeo](https://www.numbeo.com/cost-of-living/), normalizes the figures and subtracts cost of living from net salary score. The results are plotted by country, city and megahub across the world and in Europe.

```python
# imports libraries
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

## 1. Scrape data

### 1.1 Scrape Numbeo table data

```python
def get_table_data(url, id_):
    # creates object page
    page = requests.get(url)
    
    # obtains the url page's information
    soup = BeautifulSoup(page.text, 'lxml')

    # selects the table based on the id
    table = soup.find('table', id=id_)
    
    return table

def get_table_dataframe(url, id_):
    # returns the table data
    table = get_table_data(url, id_)
    
    # creates a list of headers
    headers = []
    for i in table.find_all('th'):
        title= i.text
        headers.append(title)
    # creates a new dataframe
    df = pd.DataFrame(columns = headers)
    
    # fills the new dataframe with the table data
    for i in table.find_all('tr')[1:]:
        row_data = i.find_all('td')
        row = [j.text for j in row_data]
        length = len(df)
        df.loc[length] = row
        
    return df

# creates an URL object
url = ['https://www.numbeo.com/cost-of-living/rankings.jsp', 
       'https://www.numbeo.com/cost-of-living/prices_by_city.jsp?displayCurrency=USD&itemId=105']

# creates a names object for assigning dataframes
names = ['cost', 'wage']

# assigns dataframes to names objects
for i in range(len(names)):
    globals()[names[i]] = get_table_dataframe(url[i], 't2')
```

### 1.2 Create combined dataframe

```python
# filters the wages dataframe for extra countries not in the costs dataframe
a = list(wage.sort_values('City').City)
b = list(cost.sort_values('City').City)
matching_cities = []
for i in range(len(a)):
    if(a[i] in b):
        matching_cities.append(a[i])
        
cost = cost.loc[cost.City.isin(matching_cities)].sort_values('City').reset_index(drop=True)
wage = wage.loc[wage.City.isin(matching_cities)].sort_values('City').reset_index(drop=True)

# creates a combined dataframe
df = pd.DataFrame({
    'city':list(cost['City']),
    'cost':list(cost['Cost of Living Index']),
    'wage':list(wage['Average Monthly Net Salary (After Tax)'])
})
```

## 2. Process data

### 2.1 Create simplified location labels

```python
# extracts city by dropping country and anything in brackets
df['city_labels'] = df['city'].apply(lambda x: x.split(',')[0].split('(')[0])

# extracts country by dropping city and anything in brackets
df['country_labels'] = df['city'].apply(lambda x: x.split(',')[1].split('(')[0])

# re-assigns misslablled country names for the US (initialled labbeled as their state)
df.loc[df.city.str.contains('United States'),'country_labels'] = 'US'

# re-assigns misslablled country names for the China (initialled labbeled as their province)
df.loc[df.city.str.contains('China'),'country_labels'] = 'China'
```

### 2.2 Re-assign long country names

```python
# re-assigns some country names for clairty when plotting
def re_assign(old, new):
    df.loc[df.country_labels.str.contains(old),'country_labels'] = new 

# old country names too long for plotting
old = ['United Kingdom', 'North Macedonia', 'Czech Republic', 'United Arab Emirates', 
       'Trinidad And Tobago','South Korea', 'Cayman Islands', 'Dominican Republic']

# new country names shorter for plotting
new = ['UK', 'Macedonia', 'Czechia', 'UAE', 'Trinidad', 'S. Korea', 'C. Islands', 'D. Republic']

for i in range(len(old)):
    re_assign(old[i], new[i])
```

### 2.3 Create a 'net' score  column

```python
# gets the z-scores for index values
def get_z_score(col):
    col = pd.to_numeric(col)
    avg = np.mean(col)
    sd = np.std(col)
    for i in range(len(col)):
        col[i] = (col[i]-avg)/sd
    return col

# re-assigns the cost and wages columns as their z-score  values
df['cost'] = get_z_score(df['cost'])
df['wage'] = get_z_score(df['wage'])

# creates a column of the net difference between wage and cost
df['net'] = df['wage'] - df['cost']
```

## 3. Global analysis

```python
import matplotlib.pyplot as plt

show_locations = 15

# lists popular global hubs
global_hubs = ['London', 'Paris', 'Madrid', 'Lisbon', 'Berlin', 'Rome', 
               'Amsterdam', 'Vienna', 'Prague', 'Budapest', 'Bogota',
               'Buenos Aires', 'Rio de Janeiro', 'Santiago', 'Mexico City',
               'New York', 'San Diego', 'San Francisco', 'Los Angeles', 'Miami',
               'Chicago', 'Toronto', 'Vancouver', 'Melbourne', 'Sydney', 'Auckland',
               'Wellington', 'Singapore', 'Dubai', 'Istanbul', 'Dublin']

# gets the country averages for the net column
country_avgs1 = df.groupby('country_labels')['net'].mean().sort_values(ascending = False)[0:show_locations]
country_avgs2 = df.groupby('country_labels')['net'].mean().sort_values(ascending = True)[0:show_locations]

# gets the city averages for the net column
city_avgs1 = df.sort_values('net', ascending = False)[0:show_locations]
city_avgs2 = df.sort_values('net', ascending = True)[0:show_locations]

# gets the best and worst performing global hubs
global_hubs_avgs1 = df[df['city_labels'].isin(global_hubs)].sort_values('net', ascending=False)[0:show_locations]
global_hubs_avgs2 = df[df['city_labels'].isin(global_hubs)].sort_values('net', ascending=True)[0:show_locations]

# creates a list of titles and x and y variables for plotting
title = ['Countries', 'Cities', 'Megahubs']
data1 = [['country_avgs1.index', 'country_avgs1'],['city_avgs1.city_labels', 'city_avgs1.net'],['global_hubs_avgs1.city_labels', 'global_hubs_avgs1.net']]
data2 = [['country_avgs2.index', 'country_avgs2'],['city_avgs2.city_labels', 'city_avgs2.net'],['global_hubs_avgs2.city_labels', 'global_hubs_avgs2.net']]

# gets multiple plots
def get_plots(title, data1, data2,  color):
    # creates a subplot
    graph, (plot1, plot2) = plt.subplots(2, 1, figsize=(22, 10))

    plot1.bar(eval(data1[0]), eval(data1[1]), color=color);
    plot1.set_title(title, fontsize=24);
    plot1.set_ylabel('');

    plot2.bar(eval(data2[0]), eval(data2[1]), color='darkred');
    plot2.set_title('');
    plot2.set_ylabel('');

# loops over multiple plots
for i in range(len(title)):
    get_plots(title[i], data1[i], data2[i], 'darkblue')
```
